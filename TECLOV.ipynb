# Problem 2: Funding Type Analysis 

## Part-1

```python
import numpy as np
from scipy import stats
a=master_frame[(master_frame['funding_round_type']=='angel')|(master_frame['funding_round_type']=='seed')|(master_frame['funding_round_type']=='venture')|(master_frame['funding_round_type']=='private_equity')]
a.groupby('funding_round_type')['raised_amount_usd'].agg({'funding_round_type':stats.mode})
```


### VENTURE investment type is most suitable for Teclov

## Part-2

```python
invest=master_frame[(master_frame['raised_amount_usd']>5000000)&(master_frame['raised_amount_usd']<15000000)]
x=invest[(invest['funding_round_type']=='angel')|(invest['funding_round_type']=='seed')|(invest['funding_round_type']=='venture')|(invest['funding_round_type']=='private_equity')]
x.groupby('funding_round_type')['raised_amount_usd'].agg({'funding_round_type':stats.mode})
#venture is the most suitable invesment type
```

# Problem-3 :Country Analysis

```python
#top 9 countries receiving the highest total funding
invest=master_frame[(master_frame['raised_amount_usd']>5000000)&(master_frame['raised_amount_usd']<15000000)]
venture=invest[invest['funding_round_type']=='venture']
country=venture.groupby('country_code')['raised_amount_usd'].agg(np.sum).sort_values(ascending=False).head(9)
country
```

```python
mapping=pd.read_csv('C:/Users/Administrator/Desktop/pandas/Teclov/mapping.csv')
mapping
```

```python
mapping.drop('Blanks',axis=1)
```

```python
mapping1=mapping.melt(id_vars='category_list',value_vars=['Automotive & Sports','Cleantech / Semiconductors','Entertainment','Health','Manufacturing','News, Search and Messaging','Others','Social, Finance, Analytics, Advertising'])
mapping_category=mapping1[mapping1['value']==1]
mapping_category.drop('value',axis=1)
```

# Problem-5: Sector Analysis 2

### For country-USA

```python
invest=master_frame[(master_frame['raised_amount_usd']>5000000)&(master_frame['raised_amount_usd']<15000000)]
con=invest[(invest['country_code']=='USA')&(invest['funding_round_type']=='venture')]
D1=pd.merge(con,mapping_category)
D1
```

```python
tot_investment1=D1.groupby('variable')['raised_amount_usd'].agg(np.size).sort_values(ascending=False)
tot_investment1
```

```python
company_inv=D1.groupby(['variable','Company_Name'])['Company_Name'].agg(np.size).sort_values(ascending=False)
company_inv
```

```python
tot_amount1=D1.groupby('variable').agg({'raised_amount_usd':np.sum}).sort_values(ascending=False,by='raised_amount_usd')
tot_amount1
```
